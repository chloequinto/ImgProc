```python
image = plt.imread('rose.jpg');
plt.imshow(image)

def RBGHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    red = 0.3333*R + 0.3333*G + 0.3333*B
    
    return rgb
RGBImg = RBGHue(image);
plt.imshow(RGBImg, cmap = None, alpha = 1)

plt.savefig('OGRose.png')
```

```python
image = plt.imread('rose.jpg');
plt.imshow(image)
def ImgHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    Hue = 0.33*R + 0.33*G + 0.33*B
    
    return Hue
HueImg = ImgHue(image);
plt.imshow(HueImg, cmap = None,  alpha = 1)
plt.savefig('HueRose.png')
```

```python
def PurpleHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    purple = 0.5*R + 0.5*G + 0.5*B
    
    return purple
PurpleImg = PurpleHue(image);
plt.imshow(PurpleImg, cmap = plt.cm.Purples_r,  alpha = 1)
plt.savefig('PurpleRose.png')
```

```python
def RedHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    red = 1*R + 0*G + 0*B
    
    return red
RedImg = RedHue(image);
plt.imshow(RedImg, cmap = plt.cm.Reds_r, alpha = 0.33)
plt.savefig('RedRose.png')
```

```python
def BlueHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    blue = 0*R + 0*G + 1*B
    
    return blue
BlueImg = BlueHue(image);
plt.imshow(BlueImg, cmap = plt.cm.Blues_r, alpha = 0.33)
plt.savefig('BlueRose.png')
```

```python
def GreenHue (rgb):
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    green = 0*R + 1*G + 0*B
    
    return green
GreenImg = GreenHue(image);
plt.imshow(GreenImg, cmap = plt.cm.Greens_r, alpha = 0.33)
plt.savefig('GreenRose.png')
```
