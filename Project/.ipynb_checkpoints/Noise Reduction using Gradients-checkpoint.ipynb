```python
import cv2
import sys
from matplotlib import pyplot as plt 
import numpy as np
from PIL import Image 
from skimage.io import imread
#from skimage.util import random_noise
```

```python
image = imread('Sunflower.jpg');
#img = cv2.imread('./Sunflower.jpg',0)
plt.imshow(image)
```

```python
#img = cv2.imread('Sunflower.jpg',cv2.IMREAD_GRAYSCALE)
#plt.imshow(img)

imgGrey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grey = plt.imshow(imgGrey, cmap ="gray") #change to gray scale 
#new=cv2.imwrite('grey',grey)
```

```python
I = np.zeros(image.shape, np.uint8)
mean = 0
sigma = 1
cv2.randn(I, mean, sigma) #generates random noise distribution
noisyImg = cv2.add(image,I) #adds noise to the original image
plt.imshow(noisyImg)
```
